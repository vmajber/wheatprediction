```python
import pandas as pd 
import numpy as np
from pandas.api.types import CategoricalDtype
```

```python
weather = pd.read_csv('Weatherdata.csv')
countries = ['Austria', 'Belgium', 'Bulgaria', 'Croatia', 'Cyprus', 'Czech_Republic', 'Denmark',
            'Estonia', 'Finland', 'France', 'Germany', 'Greece', 'Hungary', 'Ireland',
            'Italy', 'Latvia', 'Lithuania', 'Luxembourg', 'Malta', 'Netherlands',
            'Poland', 'Portugal', 'Romania', 'Slovakia', 'Slovenia', 'Spain', 'Sweden', 'Ukraine'] # 'Russia'
variables = ["cld", "dtr", "frs", "pet", "pre", "tmn", "tmp", "tmx", "vap", "wet"]
weather.head()
```

```python
weather = weather [~weather ["Date"].str.contains("DJF")]
weather = weather [~weather ["Date"].str.contains("MAM")]
weather = weather [~weather ["Date"].str.contains("JJA")]
weather = weather [~weather ["Date"].str.contains("SON")]
weather = weather [~weather ["Date"].str.contains("ANN")]

weather["Year"] = weather["Date"].str[0:4]
weather["Month"] = weather["Date"].str[4:7]
monthdict = {"JAN": "01", "FEB": "02", 
            'MAR':"03", "APR": "04", 
            "MAY":'05', "JUN": "06",
            "JUL": "07", "AUG": "08",
            "SEP": "09", "OCT": "10",
            "NOV": "11", "DEC": "12"}

weather = weather.replace({"Month": monthdict})
weather["Date"] = weather["Year"] + weather["Month"]
#weather = weather.drop(columns=['Month', 'Year'])
weather["Date"]= pd.to_datetime(weather["Date"], format='%Y%m')
weather = weather.sort_values(by=['Date']).reset_index(drop=True)
weather = weather.drop('Unnamed: 0', axis=1)
weather.head()
```

```python
wetvar = ["Date", 'country'] + variables
weather2 = pd.DataFrame(columns = wetvar)
for country in countries:
    weathertech = pd.DataFrame()
    weathertech["Date"] = weather["Date"]
    weathertech.loc[:,'country'] = country
    for var in variables:
        name = country+"_"+var
        weathertech[var] = weather[name]
    #weather2 = weather2.append(weathertech, ignore_index = True)
    weather2 = pd.concat([weather2, weathertech])
weather2.head()
```

```python
weather2.to_csv("visual.csv", index=False)
```

```python
dfs = []
for name, group in weather2.groupby('country'):
    for step in range(1, 9):
        for var in variables: 
            group[f'{var}_{step}'] = group[var].shift(step)
    dfs.append(group)
weatherfinal = pd.concat(dfs)
weatherfinal["Date"] = weatherfinal["Date"].shift(9)
#weatherfinal.head(12)
```

```python
weatherfinal['month'] = weatherfinal['Date'].dt.month
weatherfinal= weatherfinal[(weatherfinal['Date'] >= '1961-08-01')]
weatherfinal= weatherfinal[(weatherfinal['Date'] < '2021-01-01')]  
weatherfinal= weatherfinal.reset_index(drop=True)
weatherfinal = weatherfinal[weatherfinal['month'] == 8]
```

```python
df = pd.read_csv('wheat.csv')
df = df[["Area Code (M49)", "Area", "Year", "Value"]]
df.rename(columns={"Area Code (M49)":'Area_code'}, inplace=True)
wheat_groupped = df.groupby(['Area', 'Year'],as_index=False)[['Value']].sum()
wheat_groupped.loc[wheat_groupped["Area"] == "Czechia", "Area"] = 'Czech_Republic'
wheat_groupped["key"] = wheat_groupped['Area'] + "_" +wheat_groupped['Year'].astype(str)
wheat_groupped
```

```python
cl = wheat_groupped['Area'].unique()
df_ah  = pd.read_csv('area_harvested.csv')
df_ah = df_ah[df_ah["Area"].isin(cl)]
df_ah = df_ah[["Area Code (M49)", "Area", "Year", "Value"]]
df_ah.rename(columns={"Area Code (M49)":'Area_code'}, inplace=True)
df_ah = df_ah.groupby(['Area', 'Year'],as_index=False)[['Value']].sum()
df_ah.loc[df_ah["Area"] == "Czechia", "Area"] = 'Czech_Republic'
df_ah["key"] = df_ah['Area'] + "_" +df_ah['Year'].astype(str)
df_ah.rename(columns={'Value':'area_harvested'}, inplace=True)
df_ah.head(100000)
```

```python
joined_wheat = wheat_groupped.merge(df_ah[[ 'area_harvested','key']], how = 'left',
                left_on = 'key', right_on = 'key')
joined_wheat["target"] = joined_wheat["Value"] / joined_wheat["area_harvested"]
joined_wheat.head()
```

```python
weatherfinal["key"] = weatherfinal["country"] + "_" +weatherfinal["Date"].dt.year.astype(str)
weatherfinal
```

```python
tablefinal = weatherfinal.merge(joined_wheat[[ 'target','key']], how = 'left',
                left_on = 'key', right_on = 'key')
tablefinal
```
